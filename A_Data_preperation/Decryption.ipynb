```python
import pandas as pd
import re
# data = pd.read_csv("CSV ATTAMPT.csv")
data = pd.read_csv("nat2021us.txt")
headers_data = pd.read_csv("HEADERS ENCODING_17.12.22 V2.csv")

data.columns = ['a']
data.head(10)

data_new = pd.DataFrame()


headers_data
headers_data_filterd = headers_data[headers_data.Position.notnull()]
headers_data_filterd = headers_data_filterd[~headers_data_filterd.isin(['Position']).any(axis=1)]


only_positon = pd.DataFrame(columns=['lower', 'higher', 'name'])
data_new.head(10)
data.head(10)
headers_data_filterd
```

```python
headers_data_filterd.info()
```

```python
data_new['Birth Year'] = data['a'].str[8:12]
data_new['Birth Month'] = data['a'].str[12:14]
data_new['Time of Birth'] = data['a'].str[18:22]
data_new['Birth Day of Week'] = data['a'].str[23:31]
data_new['Birth Place'] = data['a'].str[31:32]
data_new['Reporting Flag for Birth Place'] = data['a'].str[32:33]
data_new['Facility Recode'] = data['a'].str[49:50]
data_new['Motherâ€™s Age Imputed'] = data['a'].str[72:73]
data_new['Reported Age of Mother Used Flag'] = data['a'].str[73:74]
data_new['Motherâ€™s Single Years of Age'] = data['a'].str[74:76]
data_new.head(10)
```

```python
print(headers_data_filterd.Position[0])
splited = headers_data_filterd.Position[0].partition("-")
print(splited)
arr =[]
print(int(splited[0]) - 1)
arr.append(int(splited[0]) - 1)
arr.append( int(splited[2]))
arr.append(headers_data_filterd.Description[0])
print (arr)
```

```python
for i, row in headers_data_filterd.iterrows():
  #print(row["Position"])
  if  '-' in row["Position"]:
    splited = row["Position"].partition("-")
    arr =[]
    arr.append(int(splited[0]) - 1)
    arr.append( int(splited[2]))
    arr.append(row["Description"])
    only_positon.append(arr)
    only_positon.loc[len(only_positon)] = arr
  else:
    arr =[]
    arr.append((int(row["Position"]) - 1))
    arr.append(int(row["Position"]))
    arr.append(row["Description"])
    only_positon.append(arr)
    only_positon.loc[len(only_positon)] = arr

  
only_positon
```

```python
for i in range(len(only_positon)):
  data_new[only_positon.name.loc[i]] = data['a'].str[int(only_positon.lower.loc[i]):int(only_positon.higher.loc[i])]

data_new
```
